## Remove Range

```python
temp = df_2[['date_onset_symptoms', 'date_admission_hospital']].copy()
```

```python
from dateutil import parser
import math

def func(i):
    a, b =  map(str.strip, i.split('-'))
    a = parser.parse(a)
    b = parser.parse(b)

    diff = abs(a-b)
    if diff.days > 15:
        # print('Remove range:', diff)
        return math.nan

    avg = (a if a < b else b) + diff//2
    # print(a, b, avg)
    return avg.date()

for k, z in enumerate(temp.values):
    for i, j in enumerate(z):
        if '-' in str(j) :
            temp[temp.columns[i]].iloc[k] = func(j)
```

```python
for col in temp.columns:
    temp[col] = pd.to_datetime(temp[col])
```

```python
from datetime import timedelta

def func2(date_onset_symptoms: pd.Series, date_admission_hospital: pd.Series) -> pd.Series:
    return date_onset_symptoms.isna() | date_admission_hospital.isna() | (date_onset_symptoms - date_admission_hospital < timedelta(30))

df_4 = df_3[func2(df_3.date_onset_symptoms, df_3.date_admission_hospital)].copy()
```

## Calculate Avg

```python
from datetime import timedelta

avg_df = df_4[['date_onset_symptoms', 'date_admission_hospital']].dropna().copy()

ll = list(
    map((lambda a: a.days), list(avg_df.date_onset_symptoms - avg_df.date_admission_hospital))
)

print('Avg (days):', sum(ll) // len(ll))
```

## Remove NaN

```python
from datetime import timedelta
from tqdm import tqdm

df_5 = df_4.copy()
date_onset_symptoms_isna = df_5.date_onset_symptoms.isna()
date_admission_hospital_isna = df_5.date_admission_hospital.isna()

for i in tqdm(range(len(df_5))):
    if date_onset_symptoms_isna.iloc[i]:
        df_5.date_onset_symptoms.iloc[i] = df_5.date_admission_hospital.iloc[i] - timedelta(20)
    elif date_admission_hospital_isna.iloc[i]:
        df_5.date_admission_hospital.iloc[i] = df_5.date_onset_symptoms.iloc[i] + timedelta(20)

df_5
```
